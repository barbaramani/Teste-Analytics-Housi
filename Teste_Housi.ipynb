```python
#Eliminando as colunas que tiveram registros nulos
tabela_covid = covid.drop(['Mycoplasma pneumoniae','Fio2 (venous blood gas analysis)','Urine - Nitrite','Prothrombin time (PT), Activity','D-Dimer'],axis=1)
print("Tabelas que foram eliminadas pois não possuiam registos: 'Mycoplasma pneumoniae','Fio2 (venous blood gas analysis)','Urine - Nitrite','Prothrombin time (PT), Activity','D-Dimer'")


print('Quantidade de linhas:',tabela_covid.shape[0] )
print('Quantidade de colunas:', tabela_covid.shape[1])
```

#Percentuais dos pacientes com resultado de COVID Positivo e Negativo

Quase 10% dos pacientes testaram positivo para o teste de COVID enquanto 90% testou negativo.
A seguir, mostro em gráfico essa diferença.

```python
#Mostrando os percentuais de pacientes com resultado positivo e negativo
positivo = tabela_covid[tabela_covid['SARS-Cov-2 exam result'] == 'positive']
print(len(positivo.index))
total_positivo = len(positivo.index)

negativo = tabela_covid[tabela_covid['SARS-Cov-2 exam result'] == 'negative']
print(len(negativo.index))
total_negativo = len(negativo.index)

qnt_total = tabela_covid['SARS-Cov-2 exam result'].count()
percentual_positivo = (len(positivo.index)/qnt_total) * 100
percentual_negativo = (len(negativo.index) / qnt_total) * 100

print('Percentual com resultado positivo é de: {:.2f}%'.format(percentual_positivo))
print('Percentual com resultado negativo é de: {:.2f}%'.format(percentual_negativo))
```

```python
#Mostrando graficamente

categorias = ['Positivo','Negativo']
percentuais = [percentual_positivo, percentual_negativo]

plt.bar(categorias, percentuais,color=['blue', 'pink'])
plt.xlabel('Resultados')
plt.ylabel('Percentual (%)')
plt.title ('Percentual de Resultados de Covid')

plt.show()
```

Agrupei todos os pacientes que positivaram para COVID com paciente que detectaram ou não dectaram Influenza. Note-se, que apenas 3 casos dos pacientes testaram positivo para COVID e detectaram Influenza do tipo B.


```python
#Agrupando os casos de covid e influenza tipo A e B

casos_covid = tabela_covid[(tabela_covid['SARS-Cov-2 exam result']== 'positive')]

contagem_casos = casos_covid.groupby(['SARS-Cov-2 exam result', 'Influenza A', 'Influenza B']).size().reset_index(name='Casos')
contagem_casos
```
